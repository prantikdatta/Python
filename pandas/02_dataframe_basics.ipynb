```python
#col filtering

df_sub = df[['title', 'industry', 'imdb_rating', 'language']] #subset of a dataframe i.e. only performing actions on selected cols 
```

```python
df_sub.head(5) 
```

```python
#row filtering

df[(df.release_year>=2000) & (df.release_year<=2010)]
```

```python
df[df['studio']=='Marvel Studios']
```

```python
df.info() #prints information on all the cols and their data type
```

```python
df[(df.imdb_rating == df.imdb_rating.max()) | 
    (df.imdb_rating == df.imdb_rating.min())]
```

```python
#How old is the movie

from datetime import datetime

current_year = datetime.now().year

df["age"] = df['release_year'].apply(lambda x: current_year - x)
```

```python
df.head(3)
```

```python
df["profit"] = df.apply(
    lambda x: x['revenue'] - x['budget'], axis=1
)
```
