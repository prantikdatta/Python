```python
import pandas as pd

df_sales = pd.read_excel(r'F:\CodeBasics\Python\2_pandas_basics_code\Matplotlib_seaborn_resources\linechart.xlsx')
df_sales.head()
```

## Line Chart

```python
from matplotlib import pyplot as plt

plt.figure(figsize=(12,4))
plt.plot(df_sales["Quarter"], df_sales["Fridge"], color="blue", label="Fridge")
plt.plot(df_sales["Quarter"], df_sales["Dishwasher"], color="orange", label="Dishwasher")
plt.plot(df_sales["Quarter"], df_sales["Washing Machine"], color="green", label="Washing Machine")

plt.title("Product Sales")
plt.xlabel("Quarter")
plt.ylabel("Million $")

plt.legend()

plt.show()
```

## Pie Chart

```python
total_sales = df_sales[["Fridge","Dishwasher","Washing Machine"]].sum()
total_sales
```

```python
plt.pie(total_sales, 
        labels=total_sales.index, 
        autopct='%1.1f%%', 
        startangle=140, 
        explode=(0.3,0.1,0.1), 
        shadow=True)
plt.show()
```

## Bar Chart

```python
df_sales.plot(kind="bar", x="Quarter")

plt.xticks(rotation=45)
plt.show()
```

To use Quarters as X axis, we can set them as an index on a dataframe and then use plot function

```python
df_sales_2 = df_sales.set_index("Quarter")
df_sales_2
```

```python
df_sales_2.plot(kind="bar")
plt.xticks(rotation=45)
plt.show()
```

```python
df_sales_2.plot(kind="barh")
plt.show()
```

## Histogram

```python
df_score = pd.read_excel(r'F:\CodeBasics\Python\2_pandas_basics_code\Matplotlib_seaborn_resources\histograms.xlsx')
df_score.head(3)
```
