```python
## Converting currency

def standardized_curr(curr):
    if curr == '$$' or curr == 'Dollars':
        return 'USD'
    else: 
        return curr

df_financials = pd.read_excel('2_pandas_basics_code/3_read_write_to_excel/movies_db.xlsx', sheet_name='financials', converters={
    'currency' : standardized_curr
}) 
df_financials.head(5)
```

# Write to EXCEL

## Write two dataframes to two separate sheets in excel

```python
df_stocks = pd.DataFrame({
    'tickers': ['GOOGL', 'WMT', 'MSFT'],
    'price': [845, 65, 64 ],
    'pe': [30.37, 14.26, 30.97],
    'eps': [27.82, 4.61, 2.12]
})

df_weather =  pd.DataFrame({
    'day': ['1/1/2017','1/2/2017','1/3/2017'],
    'temperature': [32,35,28],
    'event': ['Rain', 'Sunny', 'Snow']
})
```

```python
with pd.ExcelWriter('stocks_weather.xlsx') as writer:
    df_stocks.to_excel(writer, sheet_name="stocks")
    df_weather.to_excel(writer, sheet_name="weather")
```
