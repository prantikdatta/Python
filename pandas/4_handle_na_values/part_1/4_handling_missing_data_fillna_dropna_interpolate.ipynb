## FILLNA(with mean or median)

```python
df.fillna({
    'temperature' : df.temperature.mean(),
    'windspeed' : df.windspeed.mean(),
    'event' : 'No Event'    
})
```

### How?
#### ANY | ALL

```python
df.head(5)
```

```python
df.dropna(how='any') #if any of the row value is NaN then would drop it
```

```python
df.dropna(how = 'all') #the entire row needs to have NaN value
```

```python
df.dropna(thresh=2) #2 non NA value
```
