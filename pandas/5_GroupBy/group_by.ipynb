## Custom functions

create custom functions in group by having certain criteria
e.g.: temp b/w 80-90 in group A, 60-80 in group B, etc

```python
def grp (df, idx, col):
    if 80 <= df[col].loc[idx] <= 90:
        return '80-90'
    elif 60 <= df[col].loc[idx] < 80:
        return '60-79'
    else:
        return 'Others'
```

```python
g= df.groupby(lambda idx : grp(df, idx, 'temperature'))
```

```python
for key, data in g:
    print("Key : ", key)
    print("Data : ", data)
```
